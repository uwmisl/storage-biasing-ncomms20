```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import numpy as np
import seaborn as sns
import pandas as pd
import seaborn as sns
```

```python
oligo_copy_number = [5, 10, 30, 50, 100]
missing_perc = [7.4, 4.3, 2.9, 2.7, 2.5]
recov_perc = [100-m for m in missing_perc]
logical_redundancy = [round((100.0*T/(100-T)),1) for T in missing_perc]
logical_redundancy
```

```python
sns.set_style("white")
sns.set_context("notebook", font_scale=2.0, rc={"lines.linewidth": 2.5})
plt.rcParams.update({'font.size': 20})

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(8,6))
ax1.plot(oligo_copy_number, logical_redundancy, "o-", color="g")
ax1.set_xlim([0.,105])
ax1.set_xlabel("Oligo copy #")
ax1.set_ylabel('Logical redundancy (%)', color="g") 

ax2 = ax1.twinx() 
ax2.plot(oligo_copy_number, recov_perc, "o-", color="b")
ax2.set_ylabel('Recovered oligo (%)', color="b") 

ax1.set_ylim([0,10])
ax2.set_ylim([90,100])
#plt.savefig("./figs/fig_2020_02_24_logicalRedundancy_estimate.pdf", bbox_inches="tight", dpi=300, fmt="pdf")
```
